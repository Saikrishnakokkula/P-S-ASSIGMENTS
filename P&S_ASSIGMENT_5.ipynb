```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

N = 1000
calls_per_day = np.random.randint(0, 8, size=N)
unique_calls, counts = np.unique(calls_per_day, return_counts=True)
pmf = counts / N
cdf = np.cumsum(pmf)

#  PMF
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.bar(unique_calls, pmf, color='silver', edgecolor='black', alpha=0.7)
plt.title("PMF (Probability Mass Function)")
plt.xlabel("Number of Calls")
plt.ylabel("Probability")
plt.xticks(unique_calls)

# CDF
plt.subplot(1, 2, 2)
plt.step(unique_calls, cdf, where='post', color='green', alpha=0.9)
plt.title("CDF (Cumulative Distribution Function)")
plt.xlabel("Number of Calls")
plt.ylabel("Cumulative Probability")
plt.xticks(unique_calls)

plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats

N = 100
p = 0.25
k_exact = 20
k_min = 30

prob_exact_20 = stats.binom.pmf(k_exact, N, p)
prob_at_least_30 = 1 - stats.binom.cdf(k_min - 1, N, p)

expected_surgery = N * p

print(f"Probability that exactly 20 patients need surgery: {prob_exact_20:.4f}")
print(f"Probability that at least 30 patients need surgery: {prob_at_least_30:.4f}")
print(f"Expected number of patients requiring surgery: {expected_surgery:.2f}")
```
