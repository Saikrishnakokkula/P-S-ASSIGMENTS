```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


from google.colab import files
uploaded = files.upload()


car_df = pd.read_csv('/content/car_data_250.csv')


car_df.head()




plt.figure(figsize=(10, 6))
plt.hist(car_df['Price'], bins=20, color='red', edgecolor='green')
plt.title('Histogram of Car Prices')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()



# Let's print the summary statistics for the Price column
price_summary = car_df['Price'].describe()
print(price_summary)

# Additionally, we can calculate the mode of the price distribution to find the most frequent price
most_common_price_range = car_df['Price'].mode()[0]
print(f"The most common price range is around ${most_common_price_range}.")

# Alternatively, if you'd like to manually calculate the most common price range based on bins:
# Using pd.cut to create bins (ranges) for price
bins = [0, 10000, 20000, 30000, 40000, 50000, 100000]  # Example price ranges
price_labels = ['0-10k', '10k-20k', '20k-30k', '30k-40k', '40k-50k', '50k+']
car_df['Price Range'] = pd.cut(car_df['Price'], bins=bins, labels=price_labels)

# Count the frequency of cars in each price range
price_range_counts = car_df['Price Range'].value_counts()

# Display the price ranges and their counts
print("\nMost common price ranges:")
print(price_range_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (Replace 'car_data.csv' with the actual file path)
from google.colab import files
uploaded = files.upload()

df = pd.read_csv(next(iter(uploaded)))

# Inspect dataset
display(df.head())

# Scatterplot: Horsepower vs. Mileage
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Horsepower', y='Mileage', hue='Fuel Type', palette='viridis', alpha=0.7)

# Add trendline using regression plot
sns.regplot(data=df, x='Horsepower', y='Mileage', scatter=False, color='red', line_kws={'linewidth':2})

# Labels and title
plt.xlabel('Horsepower')
plt.ylabel('Mileage (MPG)')
plt.title('Horsepower vs. Mileage by Fuel Type')
plt.legend(title='Fuel Type')
plt.grid(True)

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (Replace 'car_data.csv' with the actual file path)
from google.colab import files
uploaded = files.upload()

df = pd.read_csv(next(iter(uploaded)))

# Inspect dataset
display(df.head())

# Scatterplot: Horsepower vs. Mileage
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Horsepower', y='Mileage', hue='Fuel Type', palette='viridis', alpha=0.7)

# Add trendline using regression plot
sns.regplot(data=df, x='Horsepower', y='Mileage', scatter=False, color='red', line_kws={'linewidth':2})

# Labels and title
plt.xlabel('Horsepower')
plt.ylabel('Mileage (MPG)')
plt.title('Horsepower vs. Mileage by Fuel Type')
plt.legend(title='Fuel Type')
plt.grid(True)

# Show plot
plt.show()

# Bar Chart: Number of Cars per Brand
plt.figure(figsize=(12, 6))
brand_counts = df['Brand'].value_counts()
sns.barplot(x=brand_counts.index, y=brand_counts.values, palette='coolwarm')

# Labels and title
plt.xlabel('Brand')
plt.ylabel('Number of Cars')
plt.title('Number of Cars Available for Each Brand')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Show plot
plt.show()

# Identify the brand with the highest number o
most_common_brand = brand_counts.idxmax()
print(f'The brand with the highest number of cars is: {most_common_brand}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (Replace 'car_dataset.csv' with the actual file path)
from google.colab import files
uploaded = files.upload()

df = pd.read_csv(next(iter(uploaded)))

# Ensure Year is treated as a categorical variable
df['Year'] = df['Year'].astype(str)

# Create a grouped bar chart
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Year', hue='Fuel Type', palette='Set2')

# Labels and title
plt.xlabel('Manufacturing Year')
plt.ylabel('Number of Cars')
plt.title('Distribution of Fuel Types Across Manufacturing Years')
plt.xticks(rotation=45)
plt.legend(title='Fuel Type')

# Show plot
plt.show()
```
